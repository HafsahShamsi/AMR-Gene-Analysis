```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import time

from Bio import Entrez, SeqIO
from Bio.SeqUtils import gc_fraction

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 130

# Required by NCBI
Entrez.email = "hafsahshamsi018@gmail.com"

print('All libraries imported successfully!')
```

```python
organisms = {
    'E. coli':       'NC_000913',
    'S. aureus':     'NC_002951',
    'K. pneumoniae': 'NC_009648',
    'P. aeruginosa': 'NC_002516',
    'A. baumannii':  'NC_010571',
}

records = {}
for name, accession in organisms.items():
    print(f'Fetching {name} ({accession})...')
    try:
        handle = Entrez.efetch(
            db='nucleotide',
            id=accession,
            rettype='gbwithparts',
            retmode='text'
        )
        record = SeqIO.read(handle, 'genbank')
        handle.close()
        cds_count = sum(1 for f in record.features if f.type == 'CDS')
        records[name] = record
        print(f'  ✓ {record.id} — {cds_count:,} CDS features, {len(record.seq):,} bp')
        time.sleep(0.4)  # be polite to NCBI servers
    except Exception as e:
        print(f'  ✗ Error fetching {name}: {e}')

print(f'\nSuccessfully fetched: {len(records)}/5 genomes')
```

```python
AMR_KEYWORDS = [
    'beta-lactamase', 'efflux', 'aminoglycoside', 'tetracycline',
    'resistance', 'integrase', 'transposase', 'carbapenemase',
    'fluoroquinolone', 'macrolide', 'vancomycin', 'methicillin'
]

def scan_amr_genes(record, organism_name):
    amr_genes = []
    for feature in record.features:
        if feature.type != 'CDS':
            continue
        product = feature.qualifiers.get('product', [''])[0].lower()
        gene    = feature.qualifiers.get('gene',    [''])[0].lower()
        text    = product + ' ' + gene
        for keyword in AMR_KEYWORDS:
            if keyword in text:
                seq = feature.extract(record.seq)
                amr_genes.append({
                    'organism':   organism_name,
                    'gene':       feature.qualifiers.get('gene',    ['unknown'])[0],
                    'product':    feature.qualifiers.get('product', ['unknown'])[0],
                    'keyword':    keyword,
                    'length_bp':  len(seq),
                    'gc_content': round(gc_fraction(seq) * 100, 2),
                })
                break
    return amr_genes

all_amr = []
for name, record in records.items():
    found = scan_amr_genes(record, name)
    all_amr.extend(found)
    print(f'{name:<20} → {len(found):>3} AMR-associated genes')

amr_df = pd.DataFrame(all_amr)
print(f'\nTotal AMR genes identified: {len(amr_df)}')
print(f'\nSample entries:')
amr_df.head(8)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('AMR gene distribution across ESKAPE pathogens', fontsize=14, fontweight='bold')

colors_bar = ['#E05C5C', '#5C9BE0', '#5CBE8A', '#E0A05C', '#9B5CE0']

# Chart 1: AMR genes per organism
org_counts = amr_df['organism'].value_counts()
axes[0].bar(org_counts.index, org_counts.values,
            color=colors_bar[:len(org_counts)], edgecolor='white', linewidth=1.2)
axes[0].set_title('AMR genes per organism')
axes[0].set_ylabel('Number of AMR-associated genes')
axes[0].tick_params(axis='x', rotation=20)
for i, v in enumerate(org_counts.values):
    axes[0].text(i, v + 0.5, str(v), ha='center', fontsize=10)

# Chart 2: Resistance keyword categories
keyword_counts = amr_df['keyword'].value_counts()
axes[1].barh(keyword_counts.index, keyword_counts.values,
             color='#5C9BE0', edgecolor='white')
axes[1].set_title('AMR gene categories')
axes[1].set_xlabel('Count')
axes[1].invert_yaxis()

plt.tight_layout()
plt.savefig('amr_gene_distribution.png')
plt.show()
print('Saved: amr_gene_distribution.png')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('GC content of AMR genes', fontsize=14, fontweight='bold')

palette = ['#E05C5C','#5C9BE0','#5CBE8A','#E0A05C','#9B5CE0']
org_list = amr_df['organism'].unique()

# Chart 1: GC content distribution per organism
for org, color in zip(org_list, palette):
    data = amr_df[amr_df['organism'] == org]['gc_content']
    if len(data) > 1:
        axes[0].hist(data, bins=15, alpha=0.6, label=org,
                     color=color, edgecolor='white')
axes[0].set_title('GC content distribution by organism')
axes[0].set_xlabel('GC content (%)')
axes[0].set_ylabel('Count')
axes[0].legend(fontsize=8)

# Chart 2: Gene length vs GC content scatter
for org, color in zip(org_list, palette):
    subset = amr_df[amr_df['organism'] == org]
    axes[1].scatter(subset['length_bp'], subset['gc_content'],
                    c=color, alpha=0.6, s=25, label=org)
axes[1].set_title('Gene length vs GC content')
axes[1].set_xlabel('Gene length (bp)')
axes[1].set_ylabel('GC content (%)')
axes[1].legend(fontsize=8)

plt.tight_layout()
plt.savefig('gc_content_amr.png')
plt.show()
print('Saved: gc_content_amr.png')

print('\nMean GC content by organism:')
print(amr_df.groupby('organism')['gc_content'].agg(['mean','std']).round(2))
```

```python
pivot = amr_df.pivot_table(
    index='organism',
    columns='keyword',
    values='gene',
    aggfunc='count',
    fill_value=0
)

plt.figure(figsize=(14, 5))
sns.heatmap(
    pivot,
    annot=True, fmt='d',
    cmap='YlOrRd',
    linewidths=0.5,
    cbar_kws={'label': 'Gene count'}
)
plt.title('Resistance gene profile — organisms × categories', pad=12)
plt.xlabel('Resistance category')
plt.ylabel('')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('resistance_profile.png')
plt.show()
print('Saved: resistance_profile.png')
```

```python
print('='*60)
print('AMR GENE ANALYSIS SUMMARY')
print('='*60)
print(f'Organisms analysed:      {len(records)}')
print(f'Total AMR genes found:   {len(amr_df)}')

print(f'\nBreakdown by organism:')
summary = amr_df.groupby('organism').agg(
    AMR_genes   = ('gene',       'count'),
    mean_GC     = ('gc_content', 'mean'),
    mean_len_bp = ('length_bp',  'mean')
).round(1)
print(summary.to_string())

print(f'\nTop 10 most common resistance gene products:')
print(amr_df['product'].value_counts().head(10).to_string())

amr_df.to_csv('amr_genes_summary.csv', index=False)
print(f'\nFull results saved to: amr_genes_summary.csv')
```
